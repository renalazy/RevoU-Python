## In-Class Exercise 1

Buat fungsi **untung_deposito** untuk menghitung keuntungan (return) menginvestasikan deposito setelah sekian tahun.

Keuntungan bisa dihitung sebagai:
Keuntungan=Saldo akhir - Saldo awal / Saldo awal

Misal: Keuntungan setelah menginvestasikan 7jt selama 20 tahun di deposito adalah

(14,617,064- 7,000,000)/7,000,000 = 108%

```python
def untung_deposito(saldo_akhir, saldo_awal):
    untung = (saldo_akhir - saldo_awal) / saldo_awal * 100
    return untung
```

```python
initial_balance = 7_000_000
last_balance = 14_617_064

print(f"Keuntungan yang didapat dari deposito adalah sebesar {untung_deposito(last_balance, initial_balance)}%")
```

## In-Class Exercise 3

Dalam dataset df2 ini terdapat 6 risk_grade: A, B, C, D, E, dan F.

Buat copy dataframe tersendiri untuk risk grade C, lalu tentukan:

Size of loan book
Volume of loan book
Average ticket size dari loan
Average interest rate

```python
df2 = df.copy()
```

```python
df2 = df2.rename(columns = {'grade':'risk_grade'})
```

```python
#collect data dari df2 yang mempunyai risk grade = 'C', output disimpan dalam variable baru frame_two
frame_two = df2[(df2['risk_grade'] == 'C')]
#menghitung jumlah data yang mempunya risk grade = 'C' dengan menggunakan loan_amnt untuk parameter column count, output disimpan di variable size
size = frame_two['loan_amnt'].count()
#menghitung total volume dari loan amount, mengambil column 'loan_amnt' menggunakan fungsi .sum(), output disimpan di variable volume
volume = frame_two['loan_amnt'].sum()
#menggunakan fungsi mean mengikuti arahan "in business we call", output disimpan di variable avg_ticket
avg_ticket = frame_two['loan_amnt'].mean()
#menggunakan fungsi mean mengikuti arahan "in business we call", output disimpan di variable avg_interest
avg_interest = ((frame_two['int_rate'] * volume).sum() / volume)

print(f"Size of loan book {size}")
print(f"Volume of loan book {volume}")
print(f"Average ticket size dari loan {avg_ticket}")
print(f"Average interest rate {avg_interest}")
```
